**Tutorial: Uso de OpenCV para cargar una imagen y dibujar un bounding box (caja delimitadora)**

En este tutorial, aprenderás cómo usar la biblioteca OpenCV en Python para cargar una imagen y dibujar un bounding box (caja delimitadora) en ella.



**Paso 2: Importar las bibliotecas necesarias**

Primero, necesitamos importar las bibliotecas que vamos a usar:



```python
import cv2
import matplotlib.pyplot as plt
```



**Paso 3: Cargar la imagen**

Usaremos la función `imread` de OpenCV para cargar la imagen:



```python
# Cargar la imagen
imagen = cv2.imread('/Users/eugenio/Library/CloudStorage/OneDrive-Personal/UHE/2024 Maestría IA Aplicada/Notebooks_PercepcionComputacional/YOLO-Tiempo_Real/gato2.jpg')
```

```python
# Mostramos el tamaño de la imagen
print('Tamaño de la imagen: ', imagen.shape)
```



**Paso 4: Dibujar el bounding box**

Para dibujar el bounding box, usaremos la función `rectangle` de OpenCV. Necesitamos especificar las coordenadas del punto superior izquierdo y el punto inferior derecho del bounding box. En este ejemplo, vamos a dibujar un bounding box alrededor del objeto que está en el centro de la imagen:



```python
# Definir las coordenadas del bounding box (x1, y1) y (x2, y2) en la imagen (en píxeles)
x1, y1 = 250, 200  # Punto superior izquierdo
x2, y2 = 700, 900  # Punto inferior derecho

# Dibujar el bounding box en la imagen de entrada (imagen) con el método cv2.rectangle (imagen, (x1, y1), (x2, y2), color, grosor)
imagen_rgb = cv2.rectangle(imagen, (x1, y1), (x2, y2), (0, 255, 0), 5)  # El último argumento es el grosor de la línea
```



**Paso 5: Mostrar la imagen**

Finalmente, usaremos matplotlib para mostrar la imagen con el bounding box:



```python
# Convertir la imagen de BGR a RGB
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

# Mostrar la imagen
plt.imshow(imagen_rgb)
plt.show()
```



Y eso es todo. Ahora deberías ver tu imagen con un bounding box dibujado en ella. Puedes ajustar las coordenadas del bounding box según sea necesario para que se ajuste al objeto que estás intentando delimitar.
